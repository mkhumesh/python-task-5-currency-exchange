```python
cUrl = 'https://api.coinbase.com/v2/exchange-rates'

import requests as rq
```

```python
cResp = rq.get(url=cUrl)
```

```python
cResp.content
```

```python
cDecoded = cResp.json()
```

```python
cDecoded['data']['rates']['INR']
```

```python
inrParam = {
    'currency':'INR'
}

inrReq = rq.get(url=cUrl,params=inrParam)
```

```python
inrResp = inrReq.json()
```

```python
usdRate = float(inrResp['data']['rates']['USD'])
```
